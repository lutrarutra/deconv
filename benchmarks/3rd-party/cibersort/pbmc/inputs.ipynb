```python
import os

import pandas as pd
import scanpy as sc

import DeconV as dv
```

```python
wrkdir = os.path.join("..", "..", "..", "..", "data", "pbmc")
```

```python
# Download files from NCBI GEO 'GSM4041646' and bulk sample 'GSE136148_Bulk_rawgenetable'
adata = sc.read_h5ad(os.path.join(wrkdir, "reference.h5ad"))
bulk = pd.read_table(os.path.join(wrkdir, "bulk.txt"), index_col=0)

sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)

adata = dv.tl.combine(adata, bulk)

adata.layers["counts"] = adata.X.copy()
sc.pp.normalize_total(adata, target_sum=None)
adata.layers["ncounts"] = adata.X.copy()
sc.pp.log1p(adata)
```

```python
# CIBERSORTx
# bulk
bulk.to_csv(os.path.join("data", "pbmc_bulk_data.txt"), sep="\t")
# count matrix
pd.DataFrame(adata.layers["counts"].A.T, columns=adata.obs["labels"], index=adata.var_names).to_csv(os.path.join("data", "pbmc_counts.txt"), sep="\t")
# cell types
# temp.obs[["labels", "batch"]].to_csv(os.path.join("data", "pbmc_celltypes.csv"))
```
