```python
import numpy as np
from scipy.stats import norm

from matplotlib import pyplot as plt
```

```python
x = np.linspace(-200000, 200000, 10000)
f, ax = plt.subplots(1, 1, figsize=(8,5), dpi=100)
y1 = norm.pdf(x, loc=0, scale=10000)
y2 = norm.pdf(x, loc=0, scale=100000)

ax.plot(x, y1, label="Normal(mu, s)")
ax.plot(x, y2, label="Normal(mu, 10s)")
ax.axvline(20000, color='red', linestyle='--', label="bulk gene expression")
ax.set_xticks([])
plt.legend(loc="upper left")
plt.savefig("../deconV/figures/normals.pdf", bbox_inches="tight")
```
