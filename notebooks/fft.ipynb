```python
from scipy.fft import fft, ifft, fftfreq
import numpy as np

from matplotlib import pyplot as plt
```

```python
N = 600
T = 1.0 / 600.0
x = np.linspace(start=0.0, stop=N*T, num=N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(210.0 * 2.0*np.pi*x)
yf = fft(y)
xf = fftfreq(N, T)
plt.plot(xf, 2.0/N * np.abs(yf))
```
